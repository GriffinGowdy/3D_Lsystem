```python
def Set_Voxel(xyzList, color):
    global voxelArray, colors
    
    xCoord, yCoord, zCoord = xyzList[0], xyzList[1], xyzList[2]
    voxelArray[xCoord, yCoord, zCoord] = True
    colors[xCoord, yCoord, zCoord] = color


X_DIM, Y_DIM, Z_DIM = 999, 999, 1

X_ORIGIN = X_DIM // 2
Y_ORIGIN = Y_DIM // 2
Z_ORIGIN = 0

voxelArray = np.zeros((X_DIM, Y_DIM, Z_DIM), dtype=bool)
colors = np.empty(voxelArray.shape, dtype=object)


def Read_L_System(LSystem):
    global voxelArray, colors

    if LSystem[0] == '1':
        Set_Voxel([X_ORIGIN, Y_ORIGIN, Z_ORIGIN], 'black')
    elif LSystem[0] == '0':
        Set_Voxel([X_ORIGIN, Y_ORIGIN, Z_ORIGIN], 'grey')
    
    refPos = [X_ORIGIN, Y_ORIGIN, Z_ORIGIN]
    LIFOStack = []
    
    for i in LSystem[1:]:
        if i == '1':
            refPos[1] += 1
            Set_Voxel(refPos, 'black')
        elif i == '0':
            #print('reading 0 refPos')
            #print(refPos)
            refPos[1] += 1
            #print('new placement position')
            #print(refPos)
            #print("-")
            Set_Voxel(refPos, 'grey')
        elif i == '[':
            LIFOStack.append(refPos.copy())
            #print("Appended refPos:")
            #print(refPos)
            refPos[0] += -1
            #print("New refPos:")
            #print(refPos)
            #print("-")

        elif i == ']':
            refPos = LIFOStack[-1]
            #print("Whats in the stack?")
            #print(LIFOStack)
            #print("recalling refPos")
            #print(refPos)
            #print("...")
            refPos[0] += 1
            #print("new refPos")
            #print(refPos)
            #print("-")
            LIFOStack = LIFOStack[:-1]
        else:
            print("Error: Unexpected symbol")
            break

# Sample L-system.
sampleLSystem = "11111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111111[1111111111111111111111111111111111111111111111111111111111111111[11111111111111111111111111111111[1111111111111111[11111111[1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]11111111[1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111111111111111[11111111[1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]11111111[1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]11111111111111111111111111111111[1111111111111111[11111111[1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]11111111[1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111111111111111[11111111[1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]11111111[1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111111111111111111111111111111111111111111111111111111111111111[11111111111111111111111111111111[1111111111111111[11111111[1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]11111111[1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111111111111111[11111111[1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]11111111[1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]11111111111111111111111111111111[1111111111111111[11111111[1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]11111111[1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111111111111111[11111111[1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]11111111[1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0]1111[11[1[0]0]1[0]0]11[1[0]0]1[0]0"

Read_L_System(sampleLSystem)

ax = plt.figure().add_subplot(projection='3d')
ax.voxels(voxelArray, facecolors=colors)
ax.set_axis_off()

plt.show()
```
